```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


fare_type = merged_data.groupby(['type'])
total_fare = fare_type['fare'].sum()

labels = "Rural", "Suburban", "Urban"
print(labels)
sizes = total_fare
print(sizes)
explode = (0, 0, 0.2)  # only "explode" the 3rd slice 

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, 
        radius=1.5, startangle=130, colors = ['#0FE194', '#FA95F4', '#F9C804'], textprops={'fontsize': 12})
ax1.axis('equal')
ax1.set_title("% of Total Fares by City Type")
plt.savefig('pie.png')
plt.show()
```
